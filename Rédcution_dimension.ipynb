```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Remplacer 'chemin/vers/star_num_ns.csv' par le chemin réel vers votre fichier de données.
X = pd.read_csv('/content/star_numérique_non_supervisé(1).csv')
```

```python

# Centrer et réduire les données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Créer un objet PCA et spécifier le nombre de composantes principales à conserver
n_pca =8

pca = PCA(n_components=n_pca)  # Par exemple, conserver 2 composantes principales

# Ajuster l'ACP aux données centrées et réduites
principal_components = pca.fit_transform(X_scaled)

# Créer un DataFrame avec les composantes principales
noms_colonne = ['PC'+str(i) for i in range(n_pca)]
pca_df = pd.DataFrame(data=principal_components, columns= noms_colonne)
```

```python
nom = 'dimension réuite, '+ str(n_pca) +' dimensions'
pca_df.to_csv(nom+'.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.title('ACP sur star_num_ns')
plt.show()
```
